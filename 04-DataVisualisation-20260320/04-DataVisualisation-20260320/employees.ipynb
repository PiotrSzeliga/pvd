```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
employees = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/04-DataVisualisation-20260320/employees.csv')
```

```python
employees
```

```python
employees[['age','sex']].groupby(by='sex').mean().plot(kind='bar')
```

```python
from matplotlib import colors
female, male = employees[['sex','salary']].groupby(by='sex')
#.plot(subplots=True, kind='hist', ylabel='number of employees', xlabel='salary', title='salary distribution for men and women',legend=False, layout=(2,1))
fig, (x,y) = plt.subplots(2,1)
female[1].plot(kind='hist', ax=x, legend=False, title='salary distribution of female employees', xlabel='salary of female employees', ylabel='number of female employees')
male[1].plot(kind='hist', ax=y, legend=False, title='salary distribution of male employees', xlabel='salary of male employees', ylabel='number of male employees')
fig.suptitle('Salary distribution for men and women')
plt.tight_layout()
plt.show()
```
