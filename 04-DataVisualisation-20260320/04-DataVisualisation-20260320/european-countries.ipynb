```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
countries = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/04-DataVisualisation-20260320/european-countries.csv')
```

```python
countries
```

```python
countries.query('Region == "Central Europe"').sort_values(by="Population", ascending=True)[['Name', 'Population']].plot(kind='bar', x='Name', ylabel='Number of people',title='Population of Eastern European Countires')
```

```python
countries.groupby('Region')["Area"].sum().plot(
    kind='pie',
    y="Area",
    labels=countries['Name'],
    autopct='%1.1f%%',
    figsize=(10,10),
    title='Distribution of Area by Country',
    legend=False
)
```
